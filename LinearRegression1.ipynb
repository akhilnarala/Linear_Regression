```python
from numpy import *
```

```python
x=array([0,1,2,3,4,5])
y=array([0,0.8,0.9,0.1,-0.8,-1])
```

```python
print(x)
```

```python
print(y)
```

```python
from scipy.interpolate import *
```

```python
p1=polyfit(x,y,1) 
# y = x*p[0] + p[1]   For source refer: https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.polyfit.html
```

```python
print(p1) 
# Prints slope and intercept of the linear equation
```

```python
from matplotlib.pyplot import *
```

```python
%matplotlib inline
```

```python
plot(x,y,'o')
```

```python
plot(x,y,'o')
plot(x,polyval(p1,x),'r-')
```

```python
p2=polyfit(x,y,2)
print(p2)
```

```python
p3=polyfit(x,y,3)
print(p3)
```

```python
plot(x,y,'o')
plot(x,polyval(p1,x),'r-')
plot(x,polyval(p2,x),'b--')
plot(x,polyval(p3,x),'m:')
```
